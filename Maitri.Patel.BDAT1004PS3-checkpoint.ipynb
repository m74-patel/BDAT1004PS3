```python
import pandas as pd
```

```python
users=pd.read_csv('age.csv')
users
```

Mean age per occupation

```python
users.groupby('occupation').age.mean()
```

Male ratio per occupation and sort it from the most to the least

```python
users['is_male'] = users.gender.apply(lambda x: True if x == 'M' else False)
users.is_male
```

```python
(users.groupby('occupation').is_male.sum() / users.groupby('occupation').gender.count()).sort_values(ascending = False)
```

the minimum and maximum ages

```python
print(users.groupby('occupation').age.min())
print(users.groupby('occupation').age.max())
```

Mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

 percentage of women and men in every occupation

```python
gender_occup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).count()

occup_gender = gender_occup.div(occup_count, level = "occupation")
occup_gender.loc[:, 'gender']
```

```python
import pandas as pd
import matplotlib. pyplot as mb
```

```python
us_marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us_marriage_divorce = us_marriage_divorce[
    us_marriage_divorce['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = us_marriage_divorce['Year'].values
marriages_per_capita = us_marriage_divorce['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce['Divorces_per_1000'].values
mb.bar(years-2,marriages_per_capita,width=15)
mb.bar(years-2,divorces_per_capita,width=15)
```
